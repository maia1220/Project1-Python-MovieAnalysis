```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "C:/Users/wildf/Documents/Classwork/vu-nsh-data-pt-02-2020-u-c/Project_1/clean_movie_df.csv"
movie_df = pd.read_csv(file)
# pd.set_option('display.max_columns', None)
movie_df.head()
```

```python
# First create a dataframe grouping by the 'Original_Language' column
lang_group = movie_df.groupby('original_language')

# Create a dataframe counting the number of data points for each Drug Regimen
lang_group_count = pd.DataFrame(lang_group['original_language'].count())


# Create a bar chart based off of the group series from before and label the title
count_chart = lang_group_count.plot(kind='bar', title="Count of Language Type for Each Movie", color="green")

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Language")
count_chart.set_ylabel("Count")
plt.tight_layout()

plt.savefig("LanguageDB_BarChart.png")
plt.show()
```

```python

# Scatter plot using matplotlib

x_axis=movie_df["original_language"]
y_axis=movie_df["adjusted_revenue"]

plt.scatter(x_axis, y_axis)
plt.xlabel("Language")
plt.ylabel("Adjusted Revenue")

# Create a title, x label, and y label for our chart
plt.title("95% of Database is in English, Results Inconclusive")
plt.xlabel("Language")
plt.ylabel("Adjusted Revenue")

plt.show()
```

```python
#create scatter plot 2
plt.scatter(
    movie_df['original_language'],
    movie_df['adjusted_revenue'],
    marker='o',
    facecolors='red',
    edgecolors='black',
    alpha=.75)

# Create a title, x label, and y label for our chart
plt.title("95% of Database is in English")
plt.xlabel("Language")
plt.ylabel("Adjusted Revenue")

# Save the figure
plt.savefig("ScatterLang.png")

plt.show()
```
