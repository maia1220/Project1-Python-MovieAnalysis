```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
# Load and Read File
file = "Resources/clean_movie_df_aq.csv"
movie_data = pd.read_csv(file)
movie_data.head()
```

```python
# Create a new column to reflect the release year
movie_data['release_year'] = pd.DatetimeIndex(movie_data['release_date']).year
movie_data[["title","release_year"]]
```

```python
# Reorder the list by release year and find the oldest and newest release years of the dataset
movie_data.sort_values(by=['release_year'])
```

```python
# Determine the release years of the oldest and newest movies in our dataset by finding the max and min year
oldest = movie_data['release_year'].min()
newest = movie_data['release_year'].max()
print(f"The oldest movie in the dataset was released in {oldest} and the most recent movie in the dataset was released in {newest}")
```

```python
# Create decade bins and sort movies into their corresponding decades based on release year
decade_bins = [0, 1919, 1929, 1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019]
decade_labels = ["1910s", "1920s","1930s","1940s","1950s", "1960s", "1970s", "1980s", "1990s", "2000s", "2010s"]
movie_data["movie_decade"] = pd.cut(movie_data["release_year"], decade_bins, labels=decade_labels)
movie_data[["title","movie_decade"]]
```

```python
# Create variables to isolate each movie decade
tens_data = movie_data[movie_data['movie_decade'].isin(['1910s'])]
twenties_data = movie_data[movie_data['movie_decade'].isin(['1920s'])]
thirties_data = movie_data[movie_data['movie_decade'].isin(['1930s'])]
forties_data = movie_data[movie_data['movie_decade'].isin(['1940s'])]
fifties_data = movie_data[movie_data['movie_decade'].isin(['1950s'])]
sixties_data = movie_data[movie_data['movie_decade'].isin(['1960s'])]
seventies_data = movie_data[movie_data['movie_decade'].isin(['1970s'])]
eighties_data = movie_data[movie_data['movie_decade'].isin(['1980s'])]
nineties_data = movie_data[movie_data['movie_decade'].isin(['1990s'])]
mil_data = movie_data[movie_data['movie_decade'].isin(['2000s'])]
twentytens_data = movie_data[movie_data['movie_decade'].isin(['2010s'])]
```

```python
# Find the top 5 most successful movies for each decade sorted by revenue
tens_rev = tens_data.nlargest(5, ['adjusted_revenue']) 
twenties_rev = twenties_data.nlargest(5, ['adjusted_revenue']) 
thirties_rev = thirties_data.nlargest(5, ['adjusted_revenue']) 
forties_rev = forties_data.nlargest(5, ['adjusted_revenue']) 
fifties_rev = fifties_data.nlargest(5, ['adjusted_revenue']) 
sixties_rev = sixties_data.nlargest(5, ['adjusted_revenue']) 
seventies_rev = seventies_data.nlargest(5, ['adjusted_revenue']) 
eighties_rev = eighties_data.nlargest(5, ['adjusted_revenue']) 
nineties_rev = nineties_data.nlargest(5, ['adjusted_revenue']) 
mil_rev = mil_data.nlargest(5, ['adjusted_revenue']) 
twentytens_rev = twentytens_data.nlargest(5, ['adjusted_revenue']) 

# To demonstrate, show the top 5 most successful movies of the 1970s based on revenue
seventies_rev[["title", "movie_decade","adjusted_revenue"]]
```

```python
# Create a column listing the genres for each of the top 5 movies
tens_rev_genre = tens_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
tens_rev["movie_genres"] = tens_rev_genre

twenties_rev_genre = twenties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
twenties_rev["movie_genres"] = twenties_rev_genre

thirties_rev_genre = thirties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
thirties_rev["movie_genres"] = thirties_rev_genre

forties_rev_genre = forties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
forties_rev["movie_genres"] = forties_rev_genre

fifties_rev_genre = fifties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
fifties_rev["movie_genres"] = fifties_rev_genre

sixties_rev_genre = sixties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
sixties_rev["movie_genres"] = sixties_rev_genre

seventies_rev_genre = seventies_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
seventies_rev["movie_genres"] = seventies_rev_genre

eighties_rev_genre = eighties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
eighties_rev["movie_genres"] = eighties_rev_genre

nineties_rev_genre = nineties_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
nineties_rev["movie_genres"] = nineties_rev_genre

mil_rev_genre = mil_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
mil_rev["movie_genres"] = mil_rev_genre

twentytens_rev_genre = twentytens_rev.apply(lambda x: ', '.join(x.index[x == True]), axis=1)
twentytens_rev["movie_genres"] = twentytens_rev_genre

# To confirm the column was added, show the genres for each of the top 5 movies of the 1970s
seventies_rev[["title","movie_decade","adjusted_revenue","movie_genres"]]
```

```python
# Create a new dataset containing the title and genre all of the top 5 movies for each decade
combined_decades_rev = [tens_rev, twenties_rev, thirties_rev, forties_rev, fifties_rev, sixties_rev, seventies_rev, eighties_rev, nineties_rev, mil_rev, twentytens_rev]

all_decades_rev = pd.concat(combined_decades_rev)
all_decades_rev[["title", "movie_decade","adjusted_revenue","movie_genres"]]
```
