```python
import pandas as pd
```

```python
# Load the dataset
movie_df = pd.read_csv('tmdb-movie-metadata/tmdb_5000_movies.csv')

# Preview the dataset
movie_df.head()
```

```python
# Check data types
movie_df.dtypes

'''
All numeric variables in insterest are either interger or float. 
'''
```

```python
# Check observation counts for each variable
print(movie_df.count())
```

```python
# Missing data cleaning
'''
"homepage", "overview", "tagline" will be dropped since we don't include these varaibles 
in analysis.
'''
# Make a new dataset as a copy
clean1_df=movie_df.copy()

# Delete the unused columns
clean1_df.drop(['homepage', 'overview', 'tagline'], axis=1, inplace=True)

# Drop missing data
clean2_df = clean1_df.dropna()


clean2_df.count()
# final number of observation came down to 4800. 
```

```python
# Check the summary statistics
clean2_df.describe()
```

```python
# Check missing values

zero_budget=pd.DataFrame(clean2_df.loc[clean2_df["budget"]==0])                      
valid_budget=pd.DataFrame(clean2_df.loc[clean2_df["budget"] > 0])


print(zero_budget.count()) 
# 1036 observation has a budget of $0. 

print(valid_budget["budget"].value_counts())

# multiple imputation: replace the zero values to the mean value
mean_budget= (clean2_df["budget"].mean())
clean2_df["budget"]=clean2_df["budget"].replace({0: mean_budget})

print(f'The average budget is {mean_budget}')

```

```python
# Check missing values

zero_revenue=pd.DataFrame(clean2_df.loc[clean2_df["revenue"]==0])                      
valid_revenue=pd.DataFrame(clean2_df.loc[clean2_df["revenue"] > 0])

zero_revenue.count()

# multiple imputation: replace the zero values to the mean value
mean_revenue= (clean2_df["revenue"].mean())
clean2_df["revenue"]=clean2_df["revenue"].replace({0: mean_revenue})

print(f'The average revenue is {mean_revenue}')
```

```python
# Check missing values

zero_runtime=pd.DataFrame(clean2_df.loc[clean2_df["runtime"]==0])                      
valid_runtime=pd.DataFrame(clean2_df.loc[clean2_df["runtime"] > 0])

zero_runtime.count()

# multiple imputation: replace the zero values to the mean value
mean_runtime= (clean2_df["runtime"].mean())
clean2_df["runtime"]=clean2_df["runtime"].replace({0: mean_runtime})

print(f'The average runtime is {mean_runtime}')
```

```python
# Check the summary statistics of the final dataset

clean2_df.describe()
```

```python
# Check unique identifier
clean2_df["id"].is_unique
```

```python
# Set index (run only once)

clean2_df.set_index("id", inplace=True)
```

```python
# sort index

clean2_df.sort_index(axis=0, inplace=True)
clean2_df.head()
```

```python
# Add decades
clean2_df["decade"]= " "

```

```python
# Delete the unused columns
# clean2_df.drop(['keywords', 'original_language', 'original_title', 'status'], axis=0, inplace=True)

```

```python
# Split the release_date

date_split = clean2_df["release_date"].str.split("-", expand = True)
clean2_df["release_year"] = date_split[0]
clean2_df["release_month"] = date_split[1]


clean2_df.head()
```

```python
# Inflation installation


# !pip install cpi
```

```python
# Adjuste inflation for revenue to current year

import cpi

clean2_df["release_year"]=pd.to_numeric(clean2_df["release_year"])

clean2_df['adjusted_revenue'] = clean2_df.apply(lambda x: cpi.inflate(x.revenue, x.release_year), axis=1)
clean2_df['adjusted_budget'] = clean2_df.apply(lambda x: cpi.inflate(x.budget, x.release_year), axis=1)

clean2_df.head()
```

```python

# copy to a new dataset
clean3_df=clean2_df.copy()


# Reformat the movie genres to seperate columns

movie_list = ["Action", "Adventure", "Animation", "Comedy", "Crime", "Documentary", "Drama", "Family", "Fantasy", "Foreign", "History", "Horror", "Music", "Mystery", "Romance", "Science Fiction", "Thriller", "TV Movie", "War", "Western"]

for genre in movie_list:
    clean3_df[genre]=clean3_df["genres"].str.contains(genre, regex=False)

clean3_df.head()

```

```python
clean3_df.any()
```

```python
# Sample correlation graph

import matplotlib.pyplot as plt

x_axis=clean3_df["adjusted_revenue"]
y_axis=clean3_df["popularity"]

plt.scatter(x_axis, y_axis)
plt.xlabel("Adjusted Revenue")
plt.ylabel("Popularity")
plt.show()
```

```python
# Export file as a CSV, with the Pandas index, but with the header
clean3_df.to_csv("clean_movie_df.csv", index=True, header=True)
```

Data Analysis - Ally Qi

1. The original movie title length and average vote rating 
2. The release year and average vote rating

```python
# Question 1

clean3_df[["Title Count", "vote_average"]].describe()
```

```python
# Question 1

from scipy.stats import linregress
import scipy.stats as st

x_axis=clean3_df["Title Count"]
y_axis=clean3_df["vote_average"]
plt.scatter(x_axis, y_axis)

correlation = st.pearsonr(x_axis, y_axis)
print(f"The correlation between both factors is {round(correlation[0],2)}")


(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis, y_axis)
plt.plot(x_axis,regress_values,"r-")
plt.annotate(line_eq,(40, 1),fontsize=15,color="red")
print(f"The r-squared is: {rvalue**2}")
print(f"The P value is: {pvalue}")
plt.xlabel("Original Movie Title Length")
plt.ylabel("Vote Average")
plt.title("Relationship between Title Length and Average Rating")
plt.savefig("voteavg_titleleng.png")
plt.show()
```
